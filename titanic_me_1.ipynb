```python
def impute_Age (cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
        return Age
```

```python
data_set['Age']=data_set[['Age','Pclass']].apply(impute_Age,axis=1)
```
